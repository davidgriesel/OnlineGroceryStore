# 4. Derive new variables

## 4.1. Price label

### 4.1.1. Create variable

```python
#Creating conditions
```

```python
#Condition 1
df_orders_products_merged.loc[df_orders_products_merged['prices'] <= 5, 'price_label'] = 'Low-range product'
```

```python
#Condition 2
df_orders_products_merged.loc[(df_orders_products_merged['prices'] > 5) & (df_orders_products_merged['prices'] <= 15), 'price_label'] = 'Mid-range product'
```

```python
#Condition 3
df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_label'] = 'High-range product'
```

### 4.1.2. Print frequency of new variable

```python
##Check frequency of new variable
df_orders_products_merged['price_label'].value_counts(dropna = False)
```

## 4.2. Busiest day

### 4.2.1. Create variable

```python
#Check frequency of variable values sorted by value
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Create loop and place results in list
result = []

for value in df_orders_products_merged['orders_day_of_week']:
  if value == 0:
    result.append('Busiest day')
  elif value == 4:
    result.append('Least busy')
  else:
    result.append('Regularly busy')
```

```python
#Create new variable and set it equal to results of loop
df_orders_products_merged['busiest_day'] = result
```

```python
##View list
result
```

### 4.2.2. Print frequency of new variable

```python
##Check frequency of new variable
df_orders_products_merged['busiest_day'].value_counts(dropna = False)
```

## 4.3. Busiest days

### 4.3.1. Create variable

```python
#View frequency of orders per day
df_orders_products_merged['orders_day_of_week'].value_counts(dropna = False)
```

###### The two busiest days are Saturday and Sunday, while the two slowest days are Tuesday and Wednesday

```python
#Create loop and place results in list
result_2 = []

for value in df_orders_products_merged['orders_day_of_week']:
  if value == 0 or value == 1:
    result_2.append('Busiest day')
  elif value == 3 or value == 4:
    result_2.append('Least busy')
  else:
    result_2.append('Regularly busy')
```

```python
#Create new variable and set it equal to results of loop
df_orders_products_merged['busiest_days'] = result_2
```

### 4.3.2. Print frequency of new variable

```python
##Check frequency of new variable
df_orders_products_merged['busiest_days'].value_counts(dropna = False)
```

###### Frequency of values in new variable agrees to the sum of values frequencies in orders_day_of_week variable

## 4.4. Busiest hours

### 4.4.1. Create variable

```python
#View frequency of order_hour_of_day
df_orders_products_merged['order_hour_of_day'].value_counts()
```

###### Orders are above 2m between the hours of 9am and 4pm, and less than 500 000 between 23pm and 6am.

```python
#Create loop and place results in list
result_3 = []

for value in df_orders_products_merged['order_hour_of_day']:
  if value in [10, 11, 12, 13, 14, 15]:
    result_3.append('Most orders')
  elif value in [0, 1, 2, 3, 4, 5, 6]:
    result_3.append('Fewest orders')
  else:
    result_3.append('Average orders')
```

```python
#Create new variable and set it equal to results of loop
df_orders_products_merged['busiest_hours'] = result_3
```

### 4.4.2. Print frequency of new variable

```python
##Check frequency of new variable
df_orders_products_merged['busiest_hours'].value_counts(dropna = False)
```
